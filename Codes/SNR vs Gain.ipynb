```python
import numpy as np
import matplotlib.pyplot as plt

#Frequency = 100MHz
f=100
#Wavelength = c/f = 3m
l = 3
# Power of Transmitter = 10000Watts
Pt = 10
#Tranmitter Gain = 10 dB
Gt = 1

#Radar Cross Section = 100m^2
sigma = 100
Ld=1
Gcorr=1
# Distance between Transmitter and Object = 10,000m
Rt = 10000
# Distance between Receiver and Object = 10,000m
Rr = 10000

Ft = Fr = 1
```

```python
def Pd(Gr):
    return (Pt * Gt * Ft * l * l * Gr)/(4 * np.pi * Ld * Ld * 4 * np.pi)
```

```python
def SNR(Gr):
    P = Pd(Gr)
    ratio = (Pt * Gt * l * l * Gr * sigma * Ld * Ld * Gcorr) / (Rt * Rt * 4 * np.pi * Rr * Rr * P * Gt * l * l * Gr)
    return ratio
```

```python
X=[]
Y=[]
for i in range(6,46):
    frac = i/10
    Gr = 10**frac
    ratio = SNR(Gr)
    Y.append(ratio)
    X.append(i)

plt.plot(X,Y)
plt.grid()
plt.xlabel("Reciver Gain")
plt.ylabel("SNR")
```
