```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1024                  # number of samples
Tmax = 5                  # max time (seconds)
t = np.linspace(-Tmax, Tmax, N)   # time axis
dt = t[1] - t[0]          # time step
f = np.fft.fftfreq(N, dt) # frequency bins
fshift = np.fft.fftshift(f)  # shifted so 0 freq is in center
```

```python
def plot_signal_and_ft(x, title):
    X = np.fft.fft(x)              # Fourier Transform
    Xshift = np.fft.fftshift(X)    # shift zero-freq to center

    plt.figure(figsize=(10,4))

    # Time-domain plot
    plt.subplot(1,2,1)
    plt.plot(t, x.real, 'b')
    plt.title(f"{title} (Time domain)")
    plt.xlabel("t"); plt.ylabel("x(t)")

    # Frequency-domain plot
    plt.subplot(1,2,2)
    plt.plot(fshift, np.abs(Xshift), 'r')
    plt.title(f"{title} (Frequency domain)")
    plt.xlabel("f"); plt.ylabel("|X(f)|")

    plt.tight_layout()
    plt.show()
```

```python
#delta function
x = np.zeros_like(t)
x[N//2] = 1
plot_signal_and_ft(x, "Delta δ(t)")
```

```python
#Constant Signal
x = np.ones_like(t)
plot_signal_and_ft(x, "Constant 1")
```

```python
#Complex Exponential
f0 = 5
x = np.exp(1j*2*np.pi*f0*t)
plot_signal_and_ft(x, "Complex exponential e^(j2πf0t)")
```

```python
#Cosine
x = np.cos(2*np.pi*f0*t)
plot_signal_and_ft(x, "Cosine cos(2πf0t)")
```

```python
#Rectangular Pulse
T = 2.0
x = np.where(np.abs(t) <= T/2, 1.0, 0.0)
plot_signal_and_ft(x, "Rectangular pulse")
```

```python
#Sinc Function
x = np.sinc(t)
plot_signal_and_ft(x, "Sinc(t)")
```

```python
#Gaussian
x = np.exp(-np.pi*t**2)
plot_signal_and_ft(x, "Gaussian exp(-πt²)")
```

```python
#Unit Step
x = np.where(t >= 0, 1.0, 0.0)
plot_signal_and_ft(x, "Unit step u(t)")
```
