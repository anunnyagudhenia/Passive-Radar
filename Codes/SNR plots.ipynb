```python
import numpy as np
import matplotlib.pyplot as plt
```

Changing Frequency

```python
def freq():
    Pt = 10000  # Transmitter power in Watts
    Gt = 10     # Transmitter gain
    Gr = 6      # Receiver gain
    sigma = 100 # Radar Cross Section in mÂ²
    Ld = 1      # Baseline distance factor
    Gcorr = 1   # Correlation gain
    Rt = 10000  # Distance between Transmitter and Object (m)
    Rr = 10000  # Distance between Receiver and Object (m)
    Ft = 1      # Propagation factor (transmitter to target)
    Fr = 1      # Propagation factor (target to receiver)

    X=[]
    Y=[]
    for freq in range(88, 101):
        wavelength = (3 * 10**8) / freq
        SNR = (Gt * sigma * (Ld **2) * Gcorr * Ft * (wavelength ** 2) *Gr)/((Rt**2) * (Rr **2) * ((4 * np.pi) **3) * (Ld**2))
        Y.append(SNR)
        X.append(freq)
    plt.plot(X,Y)
    plt.grid()
    plt.xlabel("Frequency")
    plt.ylabel("SNR")
freq()
```
